# QUIZ SETUP

```python
class SimpleScene:

    FONT = None

    def __init__(self, next_scene, *text):
        self.background = pygame.Surface((589, 589))
        self.background.fill((250, 250, 250))

        y = 80
        if text:
            if SimpleScene.FONT == None:
                SimpleScene.FONT = pygame.freetype.SysFont(None, 32)
            for line in text:
                SimpleScene.FONT.render_to(self.background, (120, y), line, pygame.Color('black'))
                SimpleScene.FONT.render_to(self.background, (119, y-1), line, pygame.Color('black'))
                y += 50

        self.next_scene = next_scene
        self.additional_text = None

    def start(self, text):
        self.additional_text = text

    def draw(self, screen):
        screen.blit(self.background, (0, 0))
        if self.additional_text:
            y = 180
            for line in self.additional_text:
                SimpleScene.FONT.render_to(screen, (120, y), line, pygame.Color('black'))
                SimpleScene.FONT.render_to(screen, (119, y-1), line, pygame.Color('black'))
                y += 50

    def update(self, events, dt):
        for event in events:
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_SPACE:
                    return (self.next_scene, GameState())
```

```python
class ResultScene:

    FONT = None

    def __init__(self, next_scene, *text):
        self.background = pygame.Surface((589, 589))
        self.background.fill((250, 250, 250))

        y = 80
        if text:
            if ResultScene.FONT == None:
                ResultScene.FONT = pygame.freetype.SysFont(None, 32)
            for line in text:
                ResultScene.FONT.render_to(self.background, (120, y), line, pygame.Color('black'))
                ResultScene.FONT.render_to(self.background, (119, y-1), line, pygame.Color('black'))
                y += 50

        self.next_scene = next_scene
        self.additional_text = None

    def start(self, text):
        self.additional_text = text

    def draw(self, screen):
        screen.blit(self.background, (0, 0))
        if self.additional_text:
            y = 180
            for line in self.additional_text:
                ResultScene.FONT.render_to(screen, (120, y), line, pygame.Color('black'))
                ResultScene.FONT.render_to(screen, (119, y-1), line, pygame.Color('black'))
                y += 50

    def update(self, events, dt):
        for event in events:
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_SPACE:
                    return 
```

```python
class GameState:
    def __init__(self):
        self.current_question = None
        self.right = 0
        self.wrong = 0

    def pop_question(self):
        q = random.choice(questions)
        questions.remove(q)
        self.current_question = q
        return q

    def answer(self, answer):
        if answer == self.current_question[1]:
            self.right += 1
        else:
            self.wrong += 1

    def get_result(self):
        return f'{self.right} answers correct', f'{self.wrong} answers wrong', '', 'Good!' if self.right > self.wrong else 'You can do better!'
```

```python
class GameScene:
    def __init__(self):
        self.questionCount = 3
        if SimpleScene.FONT == None:
            SimpleScene.FONT = pygame.freetype.SysFont(None, 32)

        self.rects = []
        x = 120
        y = 350
        for n in range(4):
            rect = pygame.Rect(x, y, 80, 80)
            self.rects.append(rect)
            x += 100

    def start(self, gamestate):
        self.background = pygame.Surface((589, 589))
        self.background.fill((250, 250, 250))
        self.gamestate = gamestate
        question, answer = gamestate.pop_question()
        image = pygame.image.load(os.path.join('images', question)).convert()
        self.background.blit(image, (0, 0))

    def draw(self, screen):
        screen.blit(self.background, (0, 0))
        n = 1
        for rect in self.rects:
            if rect.collidepoint(pygame.mouse.get_pos()):
                pygame.draw.rect(screen, pygame.Color('darkgrey'), rect)
            pygame.draw.rect(screen, pygame.Color('darkgrey'), rect, 5)
            SimpleScene.FONT.render_to(screen, (rect.x+30, rect.y+30), str(n), pygame.Color('black'))
            SimpleScene.FONT.render_to(screen, (rect.x+29, rect.y+29), str(n), pygame.Color('black'))
            n+=1

    def update(self, events, dt):
        for event in events:
            if event.type == pygame.MOUSEBUTTONDOWN:
                n = 1
                for rect in self.rects:
                    if rect.collidepoint(event.pos):
                        self.gamestate.answer(n)
                        if questions and self.questionCount > 1:
                            self.questionCount -= 1
                            return ('GAME', self.gamestate)
                        else:
                            return ('RESULT', self.gamestate.get_result())
                    n += 1
```

# Maze Setup

```python
def initBoard(Maze,n,a):
    background = pygame.Surface(Maze.get_size()) #create a page/board on the user's desktop 
    background = background.convert() #change the pixel format of an image
    background.fill((250, 250, 250)) #filling with grey color
    return background
```

```python
def showBoard(maze, board):
    maze.blit(board, (0, 0)) #draw one image on another
    pygame.display.flip()
```

```python
def boardPos(mouseX, mouseY, a):
    row = mouseY//a
    col = mouseX//a
    return row, col
```

```python
def drawMazeBlack(board, boardRow, boardCol,a):
    pygame.draw.rect(board, (0,0,0), (boardCol*a,boardRow*a,a,a), 0)
    return
```

```python
def drawMonster(board, boardRow, boardCol,a):
    pygame.draw.rect(board, (255,0,0), (boardCol*a,boardRow*a,a,a), 0)
    return
```

```python
def drawMove(board, boardRow, boardCol,a):
    pygame.draw.rect(board, (0,0,255), (boardCol*a,boardRow*a,a,a), 0)
    return
```

```python
def removeMove(board, boardRow, boardCol,a):
    pygame.draw.rect(board, (250,250,250), (boardCol*a,boardRow*a,a,a), 0)
    return
```

```python
def checkBlockForMonster(playerYpos, playerXpos,positions):
    if [playerYpos,playerXpos] not in positions:
        if playerYpos == 1 and playerXpos == 4:
            positions.append([playerYpos,playerXpos])
            return 1
        if playerYpos == 9 and playerXpos == 9:
            positions.append([playerYpos,playerXpos])
            return 1
        if playerYpos == 24 and playerXpos == 15:
            positions.append([playerYpos,playerXpos])
            return 1
        if playerYpos == 24 and playerXpos == 1:
            positions.append([playerYpos,playerXpos])
            return 1
        if playerYpos == 28 and playerXpos == 21:
            positions.append([playerYpos,playerXpos])
            return 1
        if playerYpos == 30 and playerXpos == 21:
            return 2
    return 0
```

```python
def display_maze(maze,n,lifelines):
    pygame.init()
    pygame.font.init()
    pygame.mixer.init()
    
    #Sounds initialized
    s = 'sounds'
    move = pygame.mixer.Sound(os.path.join(s, 'move.wav'))
    crash = pygame.mixer.Sound(os.path.join(s, 'hit.wav'))
    congrats = pygame.mixer.Sound(os.path.join(s, 'congrats.wav'))
    
    m = 1
    
    #Maze display initialized
    playerYPos = 0
    playerXPos = 15
    a = int(600/n)
    Maze = pygame.display.set_mode((a*n, a*n))
    pygame.display.set_caption('Maze')
    board = initBoard(Maze,n,a)
    running = 1    
    for i in range(0,n):
        for j in range(0,n):
            if maze[i][j]==1:
                drawMazeBlack(board,i,j,a)
    
    #Moves
    while (running == 1):
        for event in pygame.event.get():
            if event.type is pygame.QUIT:
                running = 0
            elif event.type == pygame.KEYDOWN:
                
                #Right move
                if event.key == pygame.K_RIGHT and playerXPos + 1 < 31:
                    if maze[playerYPos][playerXPos + 1] != 1: 
                        removeMove(board,playerYPos,playerXPos,a)
                        playerXPos = playerXPos + 1
                        drawMove(board,playerYPos,playerXPos,a)
                        pygame.mixer.Sound.play(move)
                
                #Down move
                if event.key == pygame.K_DOWN and playerYPos + 1 < 31:
                    if maze[playerYPos + 1][playerXPos] != 1:
                        removeMove(board,playerYPos,playerXPos,a)
                        playerYPos = playerYPos + 1
                        drawMove(board,playerYPos,playerXPos,a) 
                        pygame.mixer.Sound.play(move)
                
                #Left move
                if event.key == pygame.K_LEFT  and playerXPos - 1 < 31:
                    if maze[playerYPos][playerXPos - 1] != 1:
                        removeMove(board,playerYPos,playerXPos,a)
                        playerXPos = playerXPos - 1
                        drawMove(board,playerYPos,playerXPos,a) 
                        pygame.mixer.Sound.play(move)
                        
                #Up move        
                if event.key == pygame.K_UP and maze[playerYPos - 1][playerXPos] != 1 and playerYPos - 1 < 31:
                    if maze[playerYPos - 1][playerXPos] != 1:
                        removeMove(board,playerYPos,playerXPos,a)
                        playerYPos = playerYPos - 1
                        drawMove(board,playerYPos,playerXPos,a)
                        pygame.mixer.Sound.play(move)
                
                #Monster Hit
                if checkBlockForMonster(playerYPos, playerXPos,monster_positions) == 1:
                    pygame.mixer.Sound.play(crash)
                    result = display_quiz()                    
                    
                    if result < 3:
                        lifelines = lifelines - 3 + result
                    if lifelines < 1:
                        pygame.quit()
                        return False
                
                #Game Finished
                if checkBlockForMonster(playerYPos, playerXPos,monster_positions) == 2:
                    pygame.mixer.Sound.play(congrats)
                    displayResultBoard()
                    pygame.quit()
                    return True
                
            if(m == 0):
                running = 0
            showBoard(Maze, board)
    pygame.quit()
    return True
```
